```python
from bs4 import BeautifulSoup
import requests


base_url = "https://www.property24.co.ke"
def engine(url):
    try:
        print(f"Getting {url}")
        r = requests.get(url)
        return r
    except Exception as err:
        print(f"An error occured \n{err}")
        return None

def parse(url) -> BeautifulSoup:
    """
    Parse the html content of the page

    Args:
        url (str): the url of the page to parse

    Returns:
        BeautifulSoup: the parsed html content
    """
    driver = engine(url)
    html = driver.content
    soup = BeautifulSoup(html, 'html.parser')
    return soup

def get_links() -> list:
    """
    Get all the links on the page

    Returns:
      list: list of all the links on the page
    """
    soup = parse(base_url)
    link_tags = soup.findAll("li", class_="col-xs-3")
    links = [a.get('href') for link in link_tags for a in link.find_all('a')]
    return links

def get_listings():
    """
    Get all the listings on the page

    Returns:
        list: list of all the listings on the page
    """
    urls = get_links()
    # for url in urls:
    #     print(f"Getting {url}")
    #     soup = parse(base_url+url)
    #     print(soup)

    soup = parse(base_url+urls[0])
    print(soup)

get_listings()
# get_links()
```
